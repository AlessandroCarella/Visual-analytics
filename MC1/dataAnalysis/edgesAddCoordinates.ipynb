```python
import pandas as pd
```

```python
dfPath = '../V1/edgesClean.json'
```

```python
df = pd.read_json(dfPath)
```

```python
df
```

```python
sourceAndTargetsUnique = set(list(df["source"].unique()) + list(df["target"].unique()))
```

```python
import random

random.seed(42)

def generate_coordinates(names):
    coordinates = {}
    used_coordinates = set()  # To keep track of used coordinates

    for name in names:
        while True:
            x = random.randint(10, 1389)
            y = random.randint(10, 878)
            if (x, y) not in used_coordinates:
                used_coordinates.add((x, y))
                coordinates[name] = (x, y)
                break
    
    return coordinates
```

```python
coordinates = generate_coordinates(sourceAndTargetsUnique)
```

```python
coordinates
```

```python
import json
with open (dfPath, 'r') as f:
    data = json.load(f)
```

```python
for elem in data:
    if elem["source"] in coordinates:
        elem["sourceX"] = coordinates[elem["source"]][0]
        elem["sourceY"] = coordinates[elem["source"]][1]
    if elem["target"] in coordinates:
        elem["targetX"] = coordinates[elem["target"]][0]
        elem["targetY"] = coordinates[elem["target"]][1]
```

```python
with open ("../V1/edgesCleanWithCoordinates.json", "w") as f:
    json.dump(data, f, indent=4)
```
