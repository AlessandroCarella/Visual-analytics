```python
import pandas as pd
import json
```

```python
dfPath = "../V1/MC1.json"
```

```python
with open (dfPath, encoding="utf-8") as f:
    data = json.load(f)
```

```python
nodes = pd.json_normalize(data["nodes"])
```

```python
links = pd.json_normalize(data["links"])
```

```python
nodes
```

```python
links
```

```python
nodes.to_csv("../V1/nodes.csv", index=False)
```

```python
links.to_csv("../V1/links.csv", index=False)
```
