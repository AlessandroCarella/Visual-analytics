```python
def remove_duplicates(dicts_list):
    seen = set()
    unique_dicts = []
    for d in dicts_list:
        # Convert dictionary to tuple of items and check if it has been seen before
        t = tuple(sorted(d.items()))
        if t not in seen:
            seen.add(t)
            unique_dicts.append(d)
    return unique_dicts
```

```python
toAdd = remove_duplicates(toAdd)
```
