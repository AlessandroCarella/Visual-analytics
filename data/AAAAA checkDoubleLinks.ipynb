```python
import pandas as pd
```

```python
df = pd.read_json("mergedDatasetWithToInvestigateExtraData.json")
```

```python
dfToDict = {}
for idx, row in df.iterrows():
    source = row['source']
    target = row['target']
    typeOfLink = row['typeOfLink']
    key1 = source + " " + target
    key2 = target + " " + source
    
    if key1 not in dfToDict and key2 not in dfToDict:
        dfToDict[key1] = set()
    
    if key1 in dfToDict:
        dfToDict[key1].add(typeOfLink)
    else:
        dfToDict[key2].add(typeOfLink)
```

```python
len(dfToDict)
```

```python
# removing the 'toInvestigate' type of link cause they are made up and not in the initial dataset
KsToRemove = []
for k in dfToDict:
    if 'toInvestigate' in dfToDict[k]:
        dfToDict[k].remove('toInvestigate')
    if len (dfToDict[k]) == 0:
        KsToRemove.append(k)

for k in KsToRemove:
    del dfToDict[k]
```

```python
listDoubles = []
listTriples = []
listQuadruples = []
for k in dfToDict:
    if len (dfToDict[k]) == 2:
        listDoubles.append((k, dfToDict[k]))
    elif len (dfToDict[k]) == 3:
        listTriples.append((k, dfToDict[k]))
    elif len (dfToDict[k]) == 4:
        listQuadruples.append((k, dfToDict[k]))

len(listDoubles), len(listTriples), len(listQuadruples)
```
